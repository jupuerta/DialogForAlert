# Prueba de modelos

Primero cargo todas las librerías necesarias para la prueba de los distintos modelos.

```python
import numpy as np 
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

Cargo el dataSet de entrenamiento

```python
df = pd.read_csv('train.csv')
```

Transformo los datos para que el modelo los pueda procesar y almaceno los vectores de descriptores en una variable X y las etiquetas en una variable Y.

```python
X =  df.values[:,:-1]
Y =  df.values[:,-1]
```

## KNN

Primero separo el conjunto de entrenamiento y el conjunto de test y para ello utilizo la función train_test_split.

```python
(trainData, testData, trainLabels, testLabels) = train_test_split(X, Y,test_size=0.25, random_state=84)
(trainData, valData, trainLabels, valLabels) = train_test_split(trainData, trainLabels,test_size=0.1, random_state=84)
```

Primero voy probando con valores de k el numeros de vecinos con los cuales el algoritmo va a probar.

```python
for k in range(1,26,2):
    modelKNN = KNeighborsClassifier(n_neighbors=k)
    modelKNN.fit(trainData, trainLabels)
    score = modelKNN.score(valData,valLabels)
    print("k=%d, precisión=%.2f%%" % (k, score * 100))
```

En este modelo el que mejor precisión posee es el que posee el valor de k igual a 7.

Sin embargo, existe una alternativa al método anterior y es usar el k-fold cross validation.
Recuperamos lo primero todo nuestro conjunto de entrenamiento uniendo el conjunto de entrenamiento y el de validación.

```python
import numpy as np
trainData=np.concatenate((trainData,valData),axis=0)
trainLabels=np.concatenate((trainLabels,valLabels),axis=0)
for k in range(1,26,2):
    modelKNN = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(modelKNN,trainData,trainLabels,cv=10)
    print("k=%d, Precisión: %0.2f (+/- %0.2f)" % (k, scores.mean(), scores.std() * 2))
```

## Árbol de decisión

Lo primero que hago es probar con el hiperparámetro min_samples_split con valores en un rango de 2 a 30.

```python
for k in range(2,19,1):
    modelArbol = DecisionTreeClassifier(min_samples_split=k)
    scores = cross_val_score(modelArbol,trainData,trainLabels,cv=10) 
    print("k=%d, Precisión: %0.2f (+/- %0.2f)" % (k, scores.mean(), scores.std() * 2))
```

Como vemos el resultado que arroja no es muy revelador, por lo que, vuelvo a ejecutar el mismo proceso solo que modificando el parámetro por defecto de 'splitter' y poniendole valor 'random'

```python
for k in range(2,30,1):
    modelArbol = DecisionTreeClassifier(min_samples_split=k, splitter='random')
    scores = cross_val_score(modelArbol,trainData,trainLabels,cv=10) 
    print("k=%d, Precisión: %0.2f (+/- %0.2f)" % (k, scores.mean(), scores.std() * 2))
```

Como se ve, las precisiones, en algunos casos, han mejorado.
Sin embargo, decido probar cambiando el valor por defecto de 'criterion' y ponerle valor 'entropy'

```python
for k in range(2,30,1):
    modelArbol = DecisionTreeClassifier(min_samples_split=k, splitter='random', criterion='entropy')
    scores = cross_val_score(modelArbol,trainData,trainLabels,cv=10) 
    print("k=%d, Precisión: %0.2f (+/- %0.2f)" % (k, scores.mean(), scores.std() * 2))
```

Desgraciadamente, no se ve una mejora en el resultado.
